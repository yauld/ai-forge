```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
echo 'source $HOME/.local/bin/env' >> ~/.zshrc
source ~/.zshrc
uv --version
```

### 坑1：notebook 调用大模型时报 socksio 错误

报错场景：在 notebook 中请求模型供应商 `base_url` 时，报：

```text
Using SOCKS proxy, but the 'socksio' package is not installed
```

原因：

- `~/.zshrc` 里配置了全局代理，如 `all_proxy=socks5://127.0.0.1:1080`
- notebook 启动时会继承这些环境变量
- 底层 `httpx/openai` 发现要走 SOCKS 代理，就会额外依赖 `socksio`

解决方法：

1. 安装依赖：`uv add "httpx[socks]" `
2. 重启 notebook Kernel
3. 重新运行相关 cell
