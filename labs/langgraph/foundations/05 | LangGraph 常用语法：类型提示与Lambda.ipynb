# 05 | LangGraph常用语法：类型提示与Lambda

LangGraph常用到Python中4个类型提示与1各匿名函数，因为后面经常用到，本篇追忆一下他们的用法。

1、4个常用的类型提示:`TypedDict`、`Union`、`Optional`、`Any`，他们来自 Python 的 `typing` 模块；

2、1个匿名函数：`lambda`，它是 Python 表达式语法。

## 1. TypedDict：描述字典有哪些字段

`TypedDict` 用来给字典声明固定字段和字段类型。运行时它还是普通 `dict`，主要帮助编辑器和类型检查器理解结构。

````python
from typing import TypedDict

class User(TypedDict):
    id: int
    name: str
    active: bool

user: User = {"id": 1, "name": "Alice", "active": True}
print(user["name"])
print(type(user))
````

用法重点：适合描述 JSON、配置项、接口返回值这类“本质上就是字典”的数据。

## 2. Union：一个值可以是多种类型

`Union[int, str]` 表示这个值可以是 `int`，也可以是 `str`。Python 3.10+ 更常用 `int | str`，含义相同。

````python
from typing import Union


def format_user_id(user_id: Union[int, str]) -> str:
    return f"用户ID: {user_id}"

print(format_user_id(1001))
print(format_user_id("u-1001"))
````

但在 VS Code/Pylance 开启类型检查后，如果输入的不是要求的参数类型，虽然程序可执行，但编辑器会提示参数类型不符合 `int | str`。

同样的函数，在 Python 3.13 里可以写得更简洁：

````python
def format_order_id(order_id: int | str) -> str:
    return f"订单ID: {order_id}"

print(format_order_id(9527))
print(format_order_id("order-9527"))
````

用法重点：当参数确实允许几种明确类型时使用，不要把它当成“什么都可以”。

## 3. Optional：值可能是 None

`Optional[str]` 等价于 `str | None`，意思是：这个值要么是字符串，要么是 `None`。

````python
from typing import Optional


def get_display_name(username: Optional[str]) -> str:
    if username is None:
        return "匿名用户"
    return username


print(get_display_name("Alice"))
print(get_display_name(None))
````

用法重点：只要参数或返回值可能是 `None`，就应该明确写出来。Python 3.13 中也常写成 `str | None`。

## 4. Any：关闭类型限制

`Any` 表示这个值可以是任意类型。它很方便，但也会让类型检查变弱。

````python
from typing import Any

def debug_value(value: Any) -> None:
    print(f"值: {value!r}, 类型: {type(value).__name__}")


debug_value(123)
debug_value("hello")
debug_value({"x": 1})
````

用法重点：适合临时调试、兼容动态数据、逐步迁移旧代码。能写具体类型时，优先写具体类型。

## 5. Lambda Function：一个简短匿名函数

`lambda` 用来写简单的匿名函数。它适合只有一个表达式的小逻辑。

````python
add_tax = lambda price: round(price * 1.13, 2)

print(add_tax(100))
print(add_tax(250))
````

这句可以读成：创建一个小函数，接收 `price`，返回 `price * 1.13` 四舍五入后的结果。

它等价于：

```python
def add_tax(price):
    return round(price * 1.13, 2)
```

用法重点：`lambda 参数: 表达式` 只能写一个表达式。逻辑一复杂，就改成 `def`。

## 小结

| 名称 | 解决什么问题 |
| --- | --- |
| `TypedDict` | 描述字典字段结构 |
| `Union` 或 `int \| str` | 一个值允许多种类型 |
| `Optional` | 一个值可能是 `None` |
| `Any` | 暂时不限制类型 |
| `lambda` | 写简短匿名函数 |
